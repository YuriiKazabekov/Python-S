#### Декоратори

```python
# Структура декоратора

def decorator(funck):
    def wrapper():
        print('Start')
        funck()
        print('End') 
    return wrapper

def hellower():
    print('Hi')
    
a = decorator(hellower)
print(a)
```

```python
#### Мемонізація

def memoized(func):
    memory = {}
    def wrapper(*args, **kwargs):
        key = (args, tuple(sorted(kwargs.items())))
        if key not in memory:
            memory[key] = func(*args, **kwargs)
        return memory[key]
    return wrapper
```
